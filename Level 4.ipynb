### Merging

```python
import pandas as pd 

employees = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Name': ['Ali', 'Sara', 'Ahmed', 'Zara']
})

departments = pd.DataFrame({
    'Emp_ID': [101, 102, 103, 104],
    'Department': ['AI', 'Software', 'Network', 'AI']
})

df = employees.merge(departments, on = 'Emp_ID', how = 'inner')
df
```

```python
salary = pd.DataFrame({
    'Emp_ID':[101,102,103,104],
    'Salary':[60000,55000,50000,70000]
})

df = df.merge(salary, on = 'Emp_ID', how = 'inner')
df
```

```python
employees = pd.DataFrame({
    'Emp_ID':[101,102,103,104],
    'Name':['Ali','Sara','Ahmed','Zara']
})

salary = pd.DataFrame({
    'Emp_ID':[101,103],
    'Salary':[60000,50000]
})

print(employees.merge(salary, on = 'Emp_ID', how = 'inner'))
print(employees.merge(salary, on = 'Emp_ID', how = 'left'))
```

### Pivot Table


```python
df = pd.DataFrame({
    'Department':['AI','AI','AI','Software','Software','Network'],
    'Gender':['M','F','M','F','M','F'],
    'Salary':[60000,65000,62000,50000,52000,47000]
})

df.pivot_table(index = 'Department', columns = 'Gender', values = 'Salary', aggfunc = 'mean')
```

```python
df.pivot_table(index = 'Department', values = 'Salary', aggfunc = 'sum')
```

```python
df.pivot_table(
    index='Department',
    values='Salary',
    aggfunc=['mean', 'max', 'min']
)
```

### Melt


```python
df = pd.DataFrame({
    'Student':['Ali','Sara'],
    'Math':[90,80],
    'Physics':[85,95],
    'Chemistry':[88,91]
})

melted = df.melt(id_vars = 'Student', var_name = 'Subject', value_vars = ['Math', 'Physics', 'Chemistry'], value_name = 'Marks')

subject_order = ['Math', 'Physics', 'Chemistry']
melted['Subject'] = pd.Categorical(melted['Subject'], categories = subject_order, ordered = True)

melted = melted.sort_values(['Student', 'Subject'])
melted
```

```python
melted.rename(columns = {'Subject': 'Course', 'Marks': 'Score'}, inplace = True)
melted
```

### Working with Real Datasets


```python
df = pd.DataFrame({
    'Name':['Ali',None,'Ahmed','Sara'],
    'Age':[20,21,None,22],
    'Department':['AI',None,'AI','Software']
})

print(df.isna().sum())
df['Name'] = df['Name'].fillna('Unknown')
df['Department'] = df['Department'].fillna('Not Assigned')
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

### Final Challenge

```python
import numpy as np
employees = pd.DataFrame({
    'Emp_ID':[101,102,103,104,105],
    'Department':['AI','AI','Software','Software','Network'],
    'Salary':[60000,65000,50000,52000,47000]
})

performance = pd.DataFrame({
    'Emp_ID':[101,102,103,104,105],
    'Rating':['A','A','B','C','A']
})

df = employees.merge(performance)
df['Rating'] = df['Rating'].map({'A': 4, 'B': 3, 'C': 2})
deparment_average = df.groupby('Department')['Salary'].transform('mean')
df['High Performer'] = np.where((df['Rating'] >= 4) & (df['Salary'] > deparment_average), 'Yes', 'No')
df

```
