```python
import pandas as pd

df = pd.DataFrame({
    'Department': [
        'AI', 'AI', 'Software', 'Network',
        'Software', 'AI', 'Network', 'AI'
    ]
})

print(df['Department'].value_counts())
print(df['Department'].value_counts().idxmax())
```

```python
df = pd.DataFrame({
    'Department': ['AI', 'AI', 'Software', 'Software', 'Network', 'Network'],
    'Salary': [60000, 65000, 50000, 52000, 45000, 47000]
})

df.groupby(['Department'])['Salary'].mean()
```

```python
df.groupby(['Department'])['Salary'].aggregate( Max_Salary='max', Min_Salary ='min', Average_Salary = 'mean')
```

```python
df = pd.DataFrame({
    'Department': ['AI','AI','AI','Software','Software','Network'],
    'Gender': ['M','F','M','F','M','F'],
    'Salary': [60000,65000,62000,50000,52000,47000]
})

df.groupby(['Department','Gender'])['Salary'].mean().reset_index()
```

### Mapping


```python
df = pd.DataFrame({
    'Grade': ['A','B','A','C','B']
})

df['Points'] = df['Grade'].map({'A': 4, 'B':3, 'C':2})
df
```

```python
df = pd.DataFrame({
    'Grade': ['A','B','A','C'],
    'Credit Hours':[3, 4, 4, 2],
}, index=['Course1', 'Course2', 'Course3', 'Course4'])

df['Points'] = df['Grade'].map({'A': 4, 'B':3, 'C':2, 'D':1, 'F':0})
df['Grade Points'] = df['Points']*df['Credit Hours']

gpa = df['Grade Points'].sum()/df['Credit Hours'].sum()
print(df)
print(gpa)

```

```python
df = pd.DataFrame({
    'Status': ['Pass','Fail','Pass','Pass','Fail']
})

df['Status'] = df['Status'].replace(['Pass','Fail'],[1,0])
df
```

```python
df = pd.DataFrame({
    'Department':['AI','AI','Software','Software','Network','Network'],
    'Salary':[60000,65000,50000,52000,45000,47000]
})

df['Above_Average'] = df['Salary'] > df.groupby(['Department'])['Salary'].transform('mean')
df['Above_Average'] = df['Above_Average'].map({True: 'Yes', False: 'No'})
df

# df['Above_Average'] = (df['Salary'].gt(df.groupby('Department')['Salary'].transform('mean')).map({True:'Yes', False:'No'}))
```
