```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Ahmed', 'Zara', 'Usman'],
    'Age': [20, 22, 21, 23, 19],
    'Marks': [88, 92, 79, 85, 90]
})

df['Status'] = df['Marks'].apply(lambda x: 'Excellent' if x>=85 else 'Good')
df
```

```python
df['Pass_Fail'] = df['Marks'].apply(lambda x: 'Pass' if x >=50 else 'Fail')
df
```

```python
import numpy as np
df['Grade'] = df['Marks'].apply(lambda x: 'A' if x >= 85 else 'B' if x >= 70 and x<=84 else 'C' if x >= 50 and x<=69 else 'D')
df
```

```python
df = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Ahmed', 'Zara'],
    'Hours_Studied': [2, 5, 1, 6],
    'Marks': [40, 78, 35, 90]
})

df['Performance'] = df.apply(lambda x: 'Excellent' if x['Marks'] >= 80 and x['Hours_Studied'] >= 5 else ('Average' if x['Marks'] >= 50 else 'Poor'), axis=1)
df
```

```python
df = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Ahmed', 'Zara', 'Usman'],
    'Age': [20, 22, 21, 23, 19],
    'Marks': [88, 92, 79, 85, 90]
})

df['Scholarship'] = df.apply(lambda x: 'Yes' if x['Marks'] >= 85 and x['Age'] < 22 else 'No', axis = 1)
df
```

```python
df = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Ahmed', 'Zara', 'Usman'],
    'Attendance': [95, 80, 60, 45, 88],
    'Marks': [88, 92, 79, 85, 90]
})

df['Status'] = df.apply(lambda x: 'Top Performer' if x['Attendance'] >= 90 and x['Marks'] >= 85 else 'Good' if x['Attendance'] >= 75 else 'At Risk', axis = 1)
df
```

```python
df['Bonus_Marks'] = df.apply(lambda x: x['Marks']*1.1 if x['Marks'] >= 90 else x['Marks']*1.05 if x['Marks'] >=70 and x['Marks'] <= 89 else x['Marks'], axis = 1)
df
```
